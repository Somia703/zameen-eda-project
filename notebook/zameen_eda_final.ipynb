**Correcting column values**

**Drop Price and Area Null values rows**

```python
#remove rows with empty prices and area
df = df.dropna(subset=['Price', 'Area'])
```

**Fill missing values in string/object column values**

```python
#change empty types with "other"
df['Type'] = df['Type'].fillna('Other')

# fill missing with "Unknown"
df['Location'] = df['Location'].fillna('Unknown')
df['Description'] = df['Description'].fillna('Unknown')
```

**Correcting numerical columns**


```python
# Extract numeric part
df['area_value'] = df['Area'].str.extract(r'(\d+(\.\d+)?)', expand=False)[0].astype(float)

# Extract unit part (kanal, marla, sq yd)
df['area_unit'] = df['Area'].str.extract(r'([Kk]anal|[Mm]arla|[Ss]q\.?\s*yd)', expand=False)

# Normalize unit text
df['area_unit'] = df['area_unit'].str.lower()

# Convert area to Marla (1 kanal = 20 marla, 1 sq yd = 0.0367 marla approx)
df['area_in_marla'] = df.apply(
    lambda row:
        row['area_value'] * 20 if row['area_unit'] == 'kanal' else
        row['area_value'] * 0.0367 if row['area_unit'] == 'sq yd' else
        row['area_value'],   # already in marla
    axis=1
)
```

```python

#Correcting Prices

def clean_data(val):
  if pd.isna(val):
    return np.nan
  val = str(val).lower().replace("pkr","").replace(",","").strip()
  #print(val)

#for value in df["Area"]:
    #clean_data(value)

  if "arab" in val:
    return float(val.replace("arab", "").replace(",", "").strip()) * 1000000000
  elif "crore" in val:
    return float(val.replace("crore", "").replace(",", "").strip())* 10000000
  elif "lakh" in val:
    return float(val.replace("lakh", "").replace(",", "").strip()) * 100000
  elif "thousand" in val:
    return float(val.replace("thousand", "").replace(",", "").strip()) * 1000
  else:
    return float(val)
df["Price_value"] = df["Price"].apply(clean_data)

#Check Again
df.info()
```

```python
#Converting datatypes
#Converting to numeric and remove errors
df["Bedrooms"] = pd.to_numeric(df["Bedrooms"], errors="coerce")
df["Bathrooms"] = pd.to_numeric(df["Bathrooms"], errors="coerce")
df["Store Rooms"] = pd.to_numeric(df["Store Rooms"], errors="coerce")
df["Servant Quarters"] = pd.to_numeric(df["Servant Quarters"], errors="coerce")
df["Floors"] = pd.to_numeric(df["Floors"], errors="coerce")
df["Parking Spaces"] = pd.to_numeric(df["Parking Spaces"], errors="coerce")
df["Built in year"] = pd.to_numeric(df["Built in year"], errors="coerce")
```

 Property Type Prevalence by Top Cities


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter for top high-price cities
top_cities = ['Islamabad','Karachi', 'Lahore', 'Rawalpindi', 'Faisalabad']
city_property_data = property_data_cleaned[property_data_cleaned['City'].isin(top_cities)]

city_property_distribution = city_property_data.groupby(['City', 'Type']).agg(
    count=('Type', 'size'),
    avg_price=('Price_value', 'mean')
).reset_index()

# ---------- Plot 1: Count of property types ----------
plt.figure(figsize=(14, 8))
city_property_distribution_sorted_count = city_property_distribution.sort_values(['City', 'count'], ascending=[True, False])

sns.barplot(
    data=city_property_distribution_sorted_count,
    x='City',
    y='count',
    hue='Type'
)
plt.title('Top Property Types by Count in High-Price Cities')
plt.xlabel('City')
plt.ylabel('Number of Properties')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
